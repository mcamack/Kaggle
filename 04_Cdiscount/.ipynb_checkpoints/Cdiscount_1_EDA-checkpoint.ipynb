### Preprocessing
The original VGG model has a mean of zero for each channel, obtained by subtracting the average of each RGB channel. It also expects data in the BGR order, so we need to do some preprocessing: 

```python
vgg_mean = np.array([123.68, 116.779, 103.939]).reshape((3,1,1))

def vgg_preprocess(x):
    x = x - vgg_mean   #subtract mean
    return x[:, ::-1]  #RGB -> BGR
```

### Build VGG16 Model
Construct a VGG16 model in Keras which will accept the images from this competition as input

```python
def ConvBlock(layers, model, filters):
    for i in range(layers):
        model.add(ZeroPadding2D((1,1)))
        model.add(Conv2D(filters, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2,2), strides=(2,2), data_format="channels_first"))
```

```python
def FullyConnectedBlock(model):
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
```

```python
def VGG16():
    model = Sequential()
    model.add(Lambda(vgg_preprocess, input_shape=(3,224,224)))
    
    ConvBlock(2, model, 64)
    ConvBlock(2, model, 128)
    ConvBlock(3, model, 256)
    ConvBlock(3, model, 512)
    ConvBlock(3, model, 512)
    
    model.add(Flatten())
    FullyConnectedBlock(model)
    FullyConnectedBlock(model)
    model.add(Dense(1000, activation='softmax'))
    
    return model
```

```python
model = VGG16()
```

The following helper function will use the Keras [image.ImageDataGenerator](https://keras.io/preprocessing/image/) object with its **flow()** method to start transforming  batches of images from the directory we tell it to. It returns an Iterator which we can call with next to get the next *batch_size* amount of image/label pairs:

```python
sample_data = bson.decode_file_iter(open(r'D:\Kaggle\04_Cdiscount\train_example.bson', 'rb'))
```

```python
# Grab the next batch of images/labels

def get_batch(gen=image.ImageDataGenerator(), shuffle=True, seed=42, batch_size=batch_size):
    X_sample = np.empty((0, 3, 224, 224), dtype=np.uint8) 
    Y_sample = []

    for i in range(batch_size):
        try:
            item = next(sample_data)            
            pic = item['imgs'][0]                         #Just grab the first image for now, even if there are 2, 3, or 4
            pic_raw = imread(io.BytesIO(pic['picture']))
            #pic_resized = cv2.resize(pic_raw, (224,224))
            pic_resized = misc.imresize(pic_raw, (224,224))            
            pic_T = np.transpose(pic_resized, (2,0,1))    #get into the channels first format to match VGG16 weights for Theano
            X_sample = np.append(X_sample, [pic_T], axis=0)
            Y_sample.append(item['category_id'])
        except StopIteration:
            print('Reached last item in dataset')
            break
        
    Y_sample = np.array(Y_sample)
    return gen.flow(X_sample.astype('uint8'), Y_sample, batch_size=batch_size, shuffle=shuffle, seed=seed)
```

```python
def pred_batch(imgs):
    preds = model.predict(imgs)
    idxs = np.argmax(preds, axis=1)

    print('Shape: {}'.format(preds.shape))
    print('Predictions prob/class: ')
    
    for i in range(len(idxs)):
        idx = idxs[i]
#         print ('  {:.4f}/{}'.format(preds[i, idx], classes[idx]))
        
    cols = 4
    rows = 2
    fig, ax = plt.subplots(rows, cols, figsize=(cols*3, rows*3))
    ax = ax.ravel()
    
    for i in range(batch_size):
        ax[i].imshow(np.transpose(imgs[i], (1,2,0)))
        ax[i].set_title(str(labels[i]) + "\n" + str(preds[i, idx]) + "\n" + classes[idxs[i]])

    plt.tight_layout()
```

## Convert train.BSON into PNG files

```python
import os
```

```python
# Create subdirectories for each classification
def create_dirs(root, subdir, data):
    for i,item in enumerate(data):
        new_path = root + subdir + "\\" + str(item['category_id'])
        if not os.path.exists(new_path): os.makedirs(new_path)
```

```python
# Turn BSON images into PNG images in their respective class subdirectories
def extract_images(root, subdir, data):
    for i,item in enumerate(data):
        pic = item['imgs'][0]                         #Just grab the first image for now, even if there are 2, 3, or 4
        pic_raw = imread(io.BytesIO(pic['picture']))
        pic_resized = misc.imresize(pic_raw, (224,224))
        
        filename = root + subdir + "\\" + str(item['category_id']) + "\\" + str(item['_id']) + ".png"
        Image.fromarray(pic_resized).save(filename)
```
